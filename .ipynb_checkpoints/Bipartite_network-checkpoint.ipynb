# Davis Southern Women — Social Network Analysis

> *In 1930s Natchez, Mississippi, 18 women shaped their community one social 
> event at a time. This notebook uncovers the hidden structure of those connections.*

## About the Dataset
Collected by Davis, Gardner & Gardner and published in **Deep South (1941)**, 
this classic dataset captures observed attendance at **14 social events** by 
**18 Southern women** — one of the most studied networks in social science history.

## What This Notebook Does
-  Visualizes the **bipartite graph** — women and events as two distinct node types
-  Projects the network onto **women only** — who knew whom through shared events
- Projects the network onto **events only** — which events drew the same crowd
-  Computes **graph and node-level statistics** to uncover the social hierarchy

```python
import matplotlib.pyplot as plt
import networkx as nx
from networkx.algorithms import bipartite

G = nx.davis_southern_women_graph()
women = G.graph["top"]
clubs = G.graph["bottom"]

print("Biadjacency matrix")
print(bipartite.biadjacency_matrix(G, women, clubs))

# project bipartite graph onto women nodes
W = bipartite.projected_graph(G, women)
print()
print("#Friends, Member")
for w in women:
    print(f"{W.degree(w)} {w}")

# project bipartite graph onto women nodes keeping number of co-occurrence
# the degree computed is weighted and counts the total number of shared contacts
W = bipartite.weighted_projected_graph(G, women)
print()
print("#Friend meetings, Member")
for w in women:
    print(f"{W.degree(w, weight='weight')} {w}")

pos = nx.spring_layout(G, seed=648)  # Seed layout for reproducible node positions
nx.draw(G, pos)
plt.show()
```

# Davis Southern Women — Bipartite Graph Analysis

The initial visualization rendered all nodes in a single color, making it impossible to
distinguish women from events. To address this, the graph is redrawn **three times**, each
revealing a different layer of the social structure.

---

### Visualization Road Map

| # | Graph | Nodes | Edge meaning |
|---|-------|-------|--------------|
| 1 | Bipartite (two-color) | Women  + Events |  Woman attended event |
| 2 | Women × Women projection |  Women only | Two women co-attended an event |
| 3 | Events × Events projection |  Events only | Two events shared an attendee |

---

# Part 1 — Redrawing the Bipartite Graph with Two Colors

The original `nx.draw(G, pos)` call assigned the same color to every node, collapsing
the two-partition structure into a visually undifferentiated cloud. In this redraw, **women
are colored teal** and **events are colored coral**, making the bipartite structure — and the
boundary between the two node types — immediately visible.

---



### Import Libraries

```python
import matplotlib.pyplot as plt
import networkx as nx
from networkx.algorithms import bipartite
from matplotlib.patches import Patch
```

### Load the graph and identify the two partitions

```python
G = nx.davis_southern_women_graph()
women = G.graph["top"]
events = G.graph["bottom"]
```

### Assign Color to every node

```python
node_colors = ["#FFE0D6" if n in women else "#B2DFDB" for n in G.nodes()]
```

### Calculate node positions

```python
pos = nx.spring_layout(G, seed=648, k=0.6)
```

### Shorten the node labels

```python
labels = {n: n.split()[0] + " " + n.split()[1][0] + "." if " " in n else n
          for n in G.nodes()}
```

### Draw the graph

```python
import warnings
warnings.filterwarnings("ignore")

plt.figure(figsize=(10, 7))
nx.draw(
    G, pos,
    node_color=node_colors,
    node_size=700,
    edge_color="#455A64",
    width=0.8,
    labels=labels,        # ✅ dict of shortened names
    with_labels=True,     # ✅ boolean
    font_size=9,
    font_color="navy",
    font_weight="normal"
)

legend = [
    Patch(color="#FFE0D6", label=f"Women (n={len(women)})"),
    Patch(color="#B2DFDB", label=f"Events (n={len(events)})")
]
plt.legend(handles=legend, loc="upper left", framealpha=0.9)
plt.title("Davis Southern Women — Bipartite Graph", 
          fontsize=30,
          fontfamily="Georgia",
          fontstyle="normal",    # optional — gives it an elegant lean
          color="#0D1B2A"        # optional — dark slate instead of pure black
)

plt.tight_layout()
plt.show()
```

# Part 2 — Unipartite Southern Women's Social Network 

Here the event nodes are removed entirely. Two women are connected by an edge if they
co-attended **at least one event**. Edge thickness scales with the **number of shared events**,
so the densest lines identify the closest social ties. Node size scales with degree — larger
nodes are women embedded in more relationships.


### Import Libraries

```python
import matplotlib.pyplot as plt
import networkx as nx
from networkx.algorithms import bipartite
```

### Load Graph and Build Projection

```python
G = nx.davis_southern_women_graph()
women = G.graph["top"]
```

### Weighted projection: edge weight = number of events co-attended

```python
W = bipartite.weighted_projected_graph(G, women)
```

### Scale Node Size by Degree 

```python
 degrees = dict(W.degree())
node_sizes = [degrees[n] * 120 for n in W.nodes()]
```

### Scale Edge Width by Weight

```python
# More shared events = thicker edge
weights = [W[u][v]["weight"] for u, v in W.edges()]
max_w = max(weights)
edge_widths = [1 + 3 * (w / max_w) for w in weights]
```

### Compute Layout 

```python
pos = nx.spring_layout(W, seed=42)
```

### Draw the Graph

```python
plt.figure(figsize=(14, 10))

nx.draw_networkx_nodes(W, pos, node_color="#FFCCBC", node_size=node_sizes, alpha=0.9)
nx.draw_networkx_labels(W, pos, font_size=11, font_color="navy", font_weight="normal")
nx.draw_networkx_edges(W, pos, width=edge_widths, edge_color="#455A64", alpha=0.2)

plt.title("Unipartite Social Connections Among Southern Women",
          fontsize=30,
          fontfamily="Georgia",
          fontstyle="normal",
          color="#0D1B2A"
)
plt.axis("off")
plt.tight_layout()
plt.show()
```

### Summary Statistics

```python
print(f"Nodes: {W.number_of_nodes()}")
print(f"Edges: {W.number_of_edges()}")
print(f"Density: {nx.density(W):.3f}")
print(f"Avg clustering: {nx.average_clustering(W):.3f}")

print("\nTop 5 most-connected women:")
for name, deg in sorted(W.degree(), key=lambda x: x[1], reverse=True)[:5]:
    print(f"  {name}: {deg} connections")
```

# Part 3 -  Unipartite Social Event Connections

### Load Graph and Build Projection

```python
G = nx.davis_southern_women_graph()
events = G.graph["bottom"]
```

###  Weighted projection: edge weight = number of women who attended both events

```python
E = bipartite.weighted_projected_graph(G, events)
```

### Scale Node Size by Degree

```python
degrees = dict(E.degree())
node_sizes = [degrees[n] * 200 for n in E.nodes()]
```

### Scale Edge Width by Weight

```python
weights = [E[u][v]["weight"] for u, v in E.edges()]
max_w = max(weights)
edge_widths = [1 + 3 * (w / max_w) for w in weights]
```

### Compute Layout 

```python
pos = nx.spring_layout(E, seed=7)
```

### Draw the Graph

```python
plt.figure(figsize=(10, 7))

nx.draw_networkx_nodes(E, pos, node_color="#B2DFDB", node_size=node_sizes,
                       node_shape="s", alpha=0.9)  # squares for events
nx.draw_networkx_labels(E, pos, font_size=7, font_color="navy", font_weight="normal")
nx.draw_networkx_edges(E, pos, width=edge_widths, edge_color="#455A64", alpha=0.5)

plt.title("Events Connected Through Shared Women Attendees",
          fontsize=24,
          fontfamily="Georgia",
          fontstyle="normal",
          color="#0D1B2A"
)
plt.axis("off")
plt.tight_layout()
plt.show()
```

### Summary Statistics

```python
print(f"Nodes: {E.number_of_nodes()}")
print(f"Edges: {E.number_of_edges()}")
print(f"Density: {nx.density(E):.3f}")
print(f"Avg clustering: {nx.average_clustering(E):.3f}")

print("\nTop 5 most-connected events:")
for name, deg in sorted(E.degree(), key=lambda x: x[1], reverse=True)[:5]:
    print(f"  {name}: {deg} connections")
```

# Inferences About Relationships and Significance of Events

### Statistical Analysis of the Women's Social Group

```python
# ── Graph-Level Statistics ────────────────────────────────────────────────────
print("=" * 45)
print("   WOMEN × WOMEN NETWORK — GRAPH SUMMARY")
print("=" * 45)
print(f"  Nodes:             {W.number_of_nodes()}")
print(f"  Edges:             {W.number_of_edges()}")
print(f"  Density:           {nx.density(W):.3f}")
print(f"  Transitivity:      {nx.transitivity(W):.3f}")
print(f"  Avg clustering:    {nx.average_clustering(W):.3f}")
print(f"  Diameter:          {nx.diameter(W)}")
print(f"  Avg path length:   {nx.average_shortest_path_length(W):.3f}")
print("=" * 45)
```

## Women × Women Network — Graph-Level Analysis

The Women × Women projection reveals an extraordinarily dense and cohesive 
social network. With 18 women and 139 edges out of a possible 153, the network 
density of 0.908 indicates that nearly every woman in this community had a 
direct social connection with every other woman — a remarkably tight social 
fabric for any community.

The transitivity score of 0.928 and average clustering coefficient of 0.937 
confirm that this was not a network of isolated pairs or loose acquaintances. 
When two women both knew a third woman, there was a 92.8% chance they also 
knew each other directly — meaning social circles were almost entirely 
overlapping. Friend groups did not just intersect; they were nearly identical.

Most strikingly, the diameter of 2 and average path length of 1.092 reveal 
that this social world was virtually one degree of separation. On average, 
any two women could reach each other in just one step — they either knew 
each other directly, or shared an immediate mutual connection. This is 
consistent with a small, geographically concentrated community in 1930s 
Natchez, Mississippi, where social life revolved around a shared calendar 
of events attended by the same tight circle of women repeatedly.

```python
# ── Node-Level Statistics ─────────────────────────────────────────────────────
import pandas as pd

degree          = dict(W.degree())
weighted_degree = dict(W.degree(weight="weight"))
clustering      = nx.clustering(W)
betweenness     = nx.betweenness_centrality(W)
eigenvector     = nx.eigenvector_centrality(W)

node_stats = pd.DataFrame({
    "degree":          degree,
    "weighted_degree": weighted_degree,
    "clustering":      clustering,
    "betweenness":     betweenness,
    "eigenvector":     eigenvector
}).round(3)

node_stats = node_stats.sort_values("degree", ascending=False)
print(node_stats)
```

## Node-Level Analysis — Individual Women

The node-level statistics reveal a clear social hierarchy within this network.

At the top, seven women — Evelyn Jefferson, Ruth DeSand, Helen Lloyd, Nora 
Fayette, Sylvia Avondale, Verne Sanderson, and Theresa Anderson — each hold 
a degree of 17, meaning they were directly connected to every single other 
woman in the network. However, degree alone does not tell the full story. 
When weighted degree is examined, Theresa Anderson stands apart with a score 
of 57 — the highest in the network — meaning she not only knew everyone but 
co-attended events with them more frequently than any other woman. She was 
not just well-connected; she was deeply embedded in the social fabric.

The clustering coefficients tell an equally interesting story. Olivia Carleton, 
Flora Price, and Charlotte McDowd each have a perfect clustering score of 1.000 
— meaning every woman they knew also knew each other. At first glance this 
sounds like a sign of strong embeddedness, but paired with their low degree 
scores of 12, 12, and 11 respectively, it reveals the opposite: these women 
operated within a small, self-contained social bubble. They were peripheral 
figures whose limited connections happened to all know each other — not because 
they were central, but because their small circle was a subset of the larger 
core group.

```python
# ── Identifying Key Women in the Network ──────────────────────────────────────

# Social hub — most connections
hub = node_stats["degree"].idxmax()
print(f"\n  Social Hub:       {hub}")
print(f"  → Connected to {node_stats.loc[hub, 'degree']} women")
print(f"  → {node_stats.loc[hub, 'weighted_degree']} total shared event appearances")

# Bridge — sits between the most groups
bridge = node_stats["betweenness"].idxmax()
print(f"\n  Network Bridge:   {bridge}")
print(f"  → Betweenness score: {node_stats.loc[bridge, 'betweenness']:.3f}")
print(f"  → She connects otherwise separate social circles")

# Peripheral — least connected
peripheral = node_stats["degree"].idxmin()
print(f"\n  Most Peripheral:  {peripheral}")
print(f"  → Connected to only {node_stats.loc[peripheral, 'degree']} women")
print(f"  → Weighted degree: {node_stats.loc[peripheral, 'weighted_degree']}")
```

## Key Social Roles in the Network

**The Social Hub — Evelyn Jefferson**
Evelyn Jefferson emerges as the social hub of the network, connected to all 
17 other women with a weighted degree of 50. She attended events alongside 
virtually every member of this community repeatedly. Her presence at social 
gatherings was a near-constant — she was the connective thread that tied the 
entire group together through shared experience.

**The Network Bridge — Evelyn Jefferson**
Interestingly, Evelyn Jefferson also holds the highest betweenness centrality 
score of 0.010. While this score is low in absolute terms — which is expected 
in an extremely dense network where most women are already directly connected 
— it still identifies her as the woman most likely to sit between otherwise 
disconnected parts of the network. In a community this tightly knit, the 
bridge role is subtle but structurally meaningful.

**The Most Peripheral — Charlotte McDowd**
Charlotte McDowd sits at the outer edge of this social world with the lowest 
degree of 11 and a weighted degree of 24. She knew fewer women and co-attended 
fewer events than anyone else in the group. Her perfect clustering score of 
1.000 confirms that the women she did know all knew each other — she was not 
a bridge between groups but rather a quiet participant on the margins of an 
otherwise highly active social circle.

```python
# ── Top 3 Most Connected Women ────────────────────────────────────────────────
print("\n  Top 3 Most Connected Women:")
print("  " + "-" * 40)
top3 = node_stats.nlargest(3, "degree")
for name, row in top3.iterrows():
    print(f"  {name:<25} degree={int(row['degree'])}   "
          f"weighted={int(row['weighted_degree'])}")
```

```python
# ── Event Network Analysis ────────────────────────────────────────────────────
print("\n" + "=" * 45)
print("   EVENTS NETWORK — KEY FINDINGS")
print("=" * 45)

# Which two events share the most women (strongest edge)
strongest = max(E.edges(data=True), key=lambda x: x[2]["weight"])
print(f"\n  Strongest event pair:")
print(f"  → {strongest[0]} & {strongest[1]}")
print(f"  → {strongest[2]['weight']} women attended both")

# Which event is connected to the most other events
event_degree = dict(E.degree())
busiest_event = max(event_degree, key=event_degree.get)
print(f"\n  Most connected event:")
print(f"  → {busiest_event}")
print(f"  → Connected to {event_degree[busiest_event]} other events")

# Which event has the highest total shared attendance (weighted degree)
event_weighted = dict(E.degree(weight="weight"))
heaviest_event = max(event_weighted, key=event_weighted.get)
print(f"\n  Event with most total shared attendance:")
print(f"  → {heaviest_event}")
print(f"  → Weighted degree: {event_weighted[heaviest_event]}")

# Top 5 strongest event pairs
print(f"\n  Top 5 strongest event connections:")
print("  " + "-" * 40)
edge_list = sorted(E.edges(data=True), 
                   key=lambda x: x[2]["weight"], reverse=True)
for u, v, d in edge_list[:5]:
    print(f"  {u:<12} — {v:<12}  shared women: {d['weight']}")
```

## What the Data Reveals

**The Social Hub** — the woman with the highest degree and weighted degree
is the connective tissue of this network. She co-attended events with nearly
every other woman and did so repeatedly, making her the central figure of
1930s Southern social life in this community.

**The Bridge** — the woman with the highest betweenness centrality connects
otherwise separate social circles. Without her, parts of the network would
be disconnected. She is not necessarily the most popular, but she is the
most structurally important.

**The Periphery** — women with low degree and low weighted degree attended
fewer events and overlapped with fewer peers. Their social world was narrower
but not absent.

**The Events** — the two events with the strongest edge share the most women
in common, meaning they attracted the same crowd repeatedly. These are the
anchor events of the social calendar. Events with high weighted degree were
not just popular — they were *socially magnetic*, drawing overlapping groups
across multiple gatherings.
